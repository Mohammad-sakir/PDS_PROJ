### Reading and showing an image

Now that we have the libraries installed, lets start by reading an image.  While the first argument of the imread() method takes the image path as the input, the second argument is a flag that sepcifies how the image will be read. The two most common flags for this input are:
- IMREAD_COLOR: the specified image is loaded in color
- IMREAD_GRAYSCALE: the specified image is loaded in grayscale (we will use this later in the tutorial)

You can learn about more flag inputs [here](https://docs.opencv.org/4.5.3/d8/d6a/group__imgcodecs__flags.html).

Make sure you have the Hospital.jpg image within the tutorial folder before running the following line of code:

```python
img = cv2.imread('Hospital.jpg', cv2.IMREAD_COLOR)
```

As you can see from the code above, reading an image does not retrieve an output. Instaed it records a list of RGB values for each image pixel. Try printing img from the code above. You will see an array of arrays (that are translated as rows of each column) of these RGB values.

```python
# Reading the image array
print(img)
```

#### Displaying an image with OpenCV

This method shows the image in a separate GUI pop-up window. Look at the comments to understand the function of the lines of code. We will be using these lines throughout the tutorial to display the processed images.

#### Displaying an image with Matplotlib

This method is used to display an image by plotting the pixels from the image array.

```python
plt.imshow(img)
```

You might have noticed that the colors in the above image appear a little odd. This is because plt plots images in BGR format. This means that it maps the three numbers of each pixel with blue first, green next and then red. However, the image is encoded in RGB. This means the value for red is read as blue and the blue of blue is read as red. To turn BGR to RGB, we use the following line line of code:

```python
# converting color to RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

#Â printing using matplotlib
plt.imshow(img_rgb)
```

For this tutorial, we will use both matlabplot and OpenCV to display images. This is because while the plots will help you compare between the different outputs, the window pop-ups will show you how the image file will actually look when you save it, giving you a feel for the accurate size of the image and effects of the applied functions. ### Resizing an image

You might often come across images that are too small to observe or are too large in size which can cause pixilation. These can both lead to problems in viewing and analysing the details in the images. To overcome this probelm, you can resize the image to a size that is large enough to allow detail viewing while maintaining the quality of the image.

OpenCV has an built-in method for resizing images. In the following code, the fx and fy values determine how much larger (or smaller) the image will be in comparison to its default size using interpolation. In the code we used INTER_CUBIC interpolation which uses a cube of the neighbouring pixel to create a new interpolated pixel. You can learn about the other interpolation methods for resizing [here](https://learnopencv.com/image-resizing-with-opencv/#resize-with-interpolation).

```python
# The following line of code increases the length and heigth of the image by 0.2
# The function uses the following format: 
# resize(image_source, desired_size, horizontal_size, vertical_size, interpolation_method)
img_resize = cv2.resize(img, None, fx=1.2, fy=1.2, interpolation = cv2.INTER_CUBIC)

# converting color to RGB and printing using matplotlib
img_resize_plt = cv2.cvtColor(img_resize, cv2.COLOR_BGR2RGB)
plt.imshow(img_resize_plt) # you can see the size change in the plot by looking at the scales

# printing using OpenCV
cv2.imshow('Hospital',img_resize)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Applying filters

Filters are used to enhance an image. They can make images appear more aesthetic, but often filters are also used because they make certain features much easier to detect. Let us take a look at some of the filters we can apply using OpenCV:  

#### Grayscale

Grayscale filters can be useful when the colors in the image do not hold any special meaning. Use of bright colors that hold no meaning can act as distractors, therefore removing colors in such cases can help you focus more on the other more meaningful features of the image.

Adding a grayscale filter is very simple in OpenCV as we can simply use a built-in function to do this. To make an image appear in grayscale, we can just read the image in grayscale by using the IMREAD_GRAYSCALE flag in the imread function.

```python
# Change the image read flag to grayscale instead of color
img_gray = cv2.imread('Hospital.jpg', cv2.IMREAD_GRAYSCALE)

# converting color to RGB and printing using matplotlib
img_gray_plt  = cv2.cvtColor(img_gray, cv2.COLOR_BGR2RGB)
plt.imshow(img_gray_plt)

# printing in another window using OpenCV
cv2.imshow('Hospital',img_gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

#### Sharpen

Sharpening an image gives emphasise to the details and edges present in the image. This can be important, for example, when analysing old or scanned images. Such images often seem to fade away. Sharpening them can bring back some of the details and make it easy to detect objects within images.

Applying the sharpening filter is slightly more complex. To apply this filter, we will be using a convolution kernal with the filter2D() function. A convolution kernel is a square matrix that performs mathematical operations on each pixel of the image to achieve the desired outcome. The values of the matrix are important in determining the outcome. The filter2D() function creates a linear filter using this kernal to apply the matric operation to the pixels of the image. The convolution matrix used in the code below is from the following page:
https://learnopencv.com/image-filtering-using-convolution-in-opencv/

```python
kernel = np.array([[0, -1,  0],
                   [-1,  5, -1],
                   [0, -1,  0]])

# filter2D(image_source, outpur_image_depth, kernel_matrix)
sharp_img = cv2.filter2D(img,-1, kernel)

# converting color to RGB and printing using matplotlib
sharp_img_plt  = cv2.cvtColor(sharp_img, cv2.COLOR_BGR2RGB)
plt.imshow(sharp_img_plt)

# printing in another window using OpenCV
cv2.imshow('Sharpened Image', sharp_img)   
cv2.waitKey(0)
cv2.destroyAllWindows()
```

#### Blur/Smoothing

Blurring, which is also often known as smoothing, is used to remove details and texture from images. This is often used when you want to focus on the general appearance of the image and when the harsh details serves as distractors. 

To blur/smoothen images, we will be using the OpenCV built-in function called the GaussianBlur(). Gaussian is a popular mathematical function that is associated with a symmetric bell-shaped curve. It's symmetrical characteristic makes it very useful for image processing methods. This function performs a weighted average using a pixel's position and its distance from the center of the specified kernal (kernal size is represented in the function argument as ksize) to create a blur.

```python
# GaussianBlur(image_source, kernal_size, horizontal_standard_deviation, vertical_standard_deviation)
blur_img = cv2.GaussianBlur(img, ksize=(5,5), sigmaX=0, sigmaY=0)

# converting color to RGB and printing using matplotlib
blur_img_plt  = cv2.cvtColor(blur_img, cv2.COLOR_BGR2RGB)
plt.imshow(blur_img_plt)

# printing in another window using OpenCV
cv2.imshow('Blurred Image', blur_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

#### Vignette filter

The vignette filter is used to  add a darkened edge to images. This adds focus to the lighter part of the image. This filter can be used in images that show distracting objects at the edges and the main attribute in the centre, to draw attention to the important features.

For this filter we will, once again, be using a Gaussian function, namely, the getGaussianKernel() function. 

```python
# extracting the number of rows and columns of the image
rows = img.shape[0]
cols = img.shape[1]

# creating a vignette mask by using kernals generated by the gaussian function 
# The number determines how effect of the vignette
cols_kernel = cv2.getGaussianKernel(cols,300)
rows_kernel = cv2.getGaussianKernel(rows,300)

# creates a gaussian matix using the kernels generated above
kernel_matrix = rows_kernel * cols_kernel.T

# creating mask the vignette mask and normalising
# p.linalg.norm() is used to measure the size of the matrix
matrix_norm = np.linalg.norm(kernel_matrix)
# The formula below is retrieved from:
# https://www.geeksforgeeks.org/create-a-vignette-filter-using-python-opencv/
mask = 500 * kernel_matrix / matrix_norm
output_img = np.copy(img)

# applying the mask to the image by altering the pixel channels
for i in range(3):
    output_img[:,:,i] = output_img[:,:,i] * mask
    
# converting color to RGB and printing using matplotlib
output_img_plt  = cv2.cvtColor(output_img, cv2.COLOR_BGR2RGB)
plt.imshow(output_img_plt)
    
# printing in another window using OpenCV
cv2.imshow('Vignette Image Filter', output_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Image addition

Image addition is used when you wnt to combine two different image together. You can use image addition to:
- merge different images together by laying one on top of another
- blending two images together by adding tranparency to both

#### Overlays

Now we will look at how to add an image on top of another. This can be useful in many cases inlcuding when you need to add a watermark, cover part of an image or add text on top of an image. 

In the method below, we add a logo onto the bottom-right corner of an image by replacing the pixels of a specified area with the pixels of the logo image.

```python
bg_img = np.copy(img)
logo_img = cv2.imread('logo.png', cv2.IMREAD_COLOR)

# resizing the logo image to fit the background image appropriately 
logo_img = cv2.resize(logo_img,(300,100))

# make sure the shape of the logo image is smaller than that of the background image
print("Size of the background image: ", bg_img.shape)
print("Size of the logo image: ", logo_img.shape)

# Extracting the dimensions of the two images
bg_img_rows,bg_img_cols,bg_img_channels = bg_img.shape
logo_img_rows,logo_img_cols,logo_img_channels = logo_img.shape

# setting position on the background image
x_start = bg_img.shape[0]-logo_img[0]
y_start = bg_img.shape[1]-logo_img[1]

x_end = x_start + logo_img.shape[1]
y_end = y_start + logo_img.shape[0]

# replacing the selected pixels from the background image with pixels from the logo image
bg_img[bg_img_rows-logo_img_rows:bg_img_rows, bg_img_cols-logo_img_cols:bg_img_cols ] = logo_img

# converting color to RGB and printing using matplotlib
bg_img_plt  = cv2.cvtColor(bg_img, cv2.COLOR_BGR2RGB)
plt.imshow(bg_img_plt)

# printing in another window using OpenCV
cv2.imshow("Overlayed Image", bg_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

#### Image blending

Image blending is used to combine two images of the same size with each other by adding transparency to both the images. This is often used to add tint to images by using one block coloured image of choice with the image you want to tint. The effect of the tint depends on the amount of transparency added to the images. This technique can also be used to find common area in maps or graphs by seeing which areas overlap.

In the below example, we will be adding an ombre tint to the hospital image by adding the weight of the images together and given them both some transparency.

```python
hosp_img = np.copy(img)
cloud_img = cv2.imread('cloud.jpg', cv2.IMREAD_COLOR)

# making sure both images are the same size by converting the size of the cloud image
cloud_img = cv2.resize(cloud_img,(1000,665))

# adding the weight of the two images and giving them transparency
blended_img = cv2.addWeighted(hosp_img,0.6,cloud_img,0.4,0)

# converting color to RGB and printing using matplotlib
blended_img_plt  = cv2.cvtColor(blended_img, cv2.COLOR_BGR2RGB)
plt.imshow(blended_img_plt)

# printing in another window using OpenCV
cv2.imshow("Blended Image", blended_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Selecting parts of an image

Throughout this tutorial, we talked about several ways of giving focus to the important features of an image and finding ways to remove focus from the distracting features. One of the best way you can do this, especially if the distracting features hold no information that is important to you, is by discarding the distracting features altogether. We will learn to do this through two different ways:

1) Cropping

2) Background removal

#### Cropping

Cropping an image to remove unnecessary details is commonly used. This method allows us to discard the un-important part of the image and allows you to focus on the more important part.

The method we use below to crop an image basically calls a subset of the array containing the image pixels that cover my desired area of the image.

```python
# selecting a subset of the pixels 
cropped_img = img[30:1000, 520:1000]

# converting color to RGB and printing using matplotlib
cropped_img_plt  = cv2.cvtColor(cropped_img , cv2.COLOR_BGR2RGB)
plt.imshow(cropped_img_plt)

# printing in another window using OpenCV
cv2.imshow("Cropped Image", cropped_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

#### Removing background

We often want to remove the background of an image when the background is distracting or when we want to place the main feature of the image on top of another image without showing the background to creating a more visually appealing outcome.

In the following set of codes, we will be working on the latter by repeating the steps mentioned above, for overlaying the Alberta Health Service logo on top of the hospital image. But this time, by making the outcome more aesthetically pleasing by removing the black background of the logo first.

We will do this by creating two masks for the Alberta Health Service logo image. One mask will contain the text outlines with the background blacked-out, while the other will contain the text outlines with the foreground blacked-out. We will use these masks to create a place for the foreground of the logo on the hospital image and extract the texts from the logo and place it onto the hospital image. Follow the comments below to get a better step-by-step understanding of how we do this.

```python
bg_img = np.copy(img)
logo_img = np.copy(logo_img)

# Extracting the dimensions of the two images
bg_img_rows,bg_img_cols,bg_img_channels = bg_img.shape
logo_img_rows,logo_img_cols,logo_img_channels = logo_img.shape

# Idensitying the locatioon on the background image
area = bg_img[bg_img_rows-logo_img_rows:bg_img_rows, bg_img_cols-logo_img_cols:bg_img_cols ]

# Creating a mask for the logo image
logo_img_gray = cv2.cvtColor(logo_img,cv2.COLOR_BGR2GRAY)
ret, mask = cv2.threshold(logo_img_gray, 10, 255, cv2.THRESH_BINARY)

# Creating an inverse mask for the logo
mask_inv = cv2.bitwise_not(mask)

# Apply the inverve mask on the background image to change color of that area to black
bg_img_bg = cv2.bitwise_and(area,area,mask = mask_inv)

# Taking the mask of the logo image
logo_img_fg = cv2.bitwise_and(logo_img,logo_img,mask = mask)

# Place the selected part from the mogo image and apply it to the selected area of the backgorund image
selected_logo = cv2.add(bg_img_bg,logo_img_fg)
bg_img[bg_img_rows-logo_img_rows:bg_img_rows, bg_img_cols-logo_img_cols:bg_img_cols ] = selected_logo


# converting color to RGB and printing using matplotlib
no_bg_logo_plt  = cv2.cvtColor(bg_img , cv2.COLOR_BGR2RGB)
plt.imshow(no_bg_logo_plt)

# printing in another window using OpenCV
cv2.imshow("No Background Logo Image", bg_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

To create an inverse effect by showing the backgound of the logo image while making the text area transparent, we use the same code as above, but just switch the masks for bg_img_bg and logo_img_fg.

```python
bg_img = np.copy(img)
logo_img = np.copy(logo_img)

# Extracting the dimensions of the two images
bg_img_rows,bg_img_cols,bg_img_channels = bg_img.shape
logo_img_rows,logo_img_cols,logo_img_channels = logo_img.shape

# Idensitying the locatioon on the background image
area = bg_img[bg_img_rows-logo_img_rows:bg_img_rows, bg_img_cols-logo_img_cols:bg_img_cols ]

# Creating a mask for the logo image
logo_img_gray = cv2.cvtColor(logo_img,cv2.COLOR_BGR2GRAY)
ret, mask = cv2.threshold(logo_img_gray, 10, 255, cv2.THRESH_BINARY)

# Creating an inverse mask for the logo
mask_inv = cv2.bitwise_not(mask)

# Apply the inverve mask on the background image to change color of that area to black
bg_img_bg = cv2.bitwise_and(area,area,mask = mask)

# Taking the mask of the logo image
logo_img_fg = cv2.bitwise_and(logo_img,logo_img,mask = mask_inv)

# Place the selected part from the mogo image and apply it to the selected area of the backgorund image
selected_logo = cv2.add(bg_img_bg,logo_img_fg)
bg_img[bg_img_rows-logo_img_rows:bg_img_rows, bg_img_cols-logo_img_cols:bg_img_cols ] = selected_logo


# converting color to RGB and printing using matplotlib
no_bg_logo_plt  = cv2.cvtColor(bg_img , cv2.COLOR_BGR2RGB)
plt.imshow(no_bg_logo_plt)

# printing in another window using OpenCVga
cv2.imshow("No Background Logo Image", bg_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
