# 1D Data

## 1. Bar Plots/Charts

If your data is **1D and categorical**, the only real option for visualizing this data is through some kind of bar chart (though variants like stacked bar charts are also possible). 

In the following example, we will visualize the total Revenue by Month. 
The month is not a numeric value, it is rather an ordinal (an ordered categorical variable)

```python
# Example 1 - Total Revenue by Month

ax = sns.barplot(x="Month_ID", y="Revenue", data=raw_data)



# Notes: 
# 1 - the lines signify the confidence interval: it simply computes the percentile limits. 
# This interval has a defined mean of 50 (median) and a default range of 95% confidence interval. 
```

So, what seaborn has done here, is that it grouped the entries from the dataframe by month ID, and then it took the mean revenue for each group.

we can check this out by:

```python
raw_data.groupby('Month_ID').Revenue.mean()
```

So, these numbers here are what is displayed in the bar plot above.

The barplot below uses the marketing data set. It shows the Total Revenue per Month. Its uses the blues palette, which has variations of the color blue.

```python
# Total Revenue by Month

ax = sns.barplot(x="Month_ID", y="Revenue", data=raw_data, palette = 'Blues')
```

```python
# Example 2 - Total Revenue by Month - Remove the Confidence Interval

ax = sns.barplot(x="Month_ID", y="Revenue", data=raw_data, palette='Greens', ci=None)
```

```python
# In matplotlib
fig, ax= plt.subplots(figsize=(8,6))
# fig = plt.figure(figsize=(8,6))
# ax = fig.add_subplot(1,1,1)

ax.bar(raw_data['Month_ID'], raw_data["Revenue"])
ax.set_xlabel('Month')
ax.set_ylabel('Revenue')
```

```python
# Example 3 - Total Revenue by Month - Remove the Confidence Interval - By Promo
ax = sns.barplot(x="Month_ID", y="Revenue", data=raw_data, ci=False, hue = 'Promo')
```

```python
# Example 4 - Total Revenue by Month - Remove the Confidence Interval - By Promo - Changing direction
ax = sns.barplot(x="Revenue", y="Year", ci=False, hue = 'Promo', orient = 'h', data=raw_data)
```

## 2. Histograms â€” numerical data

For **1D numerical data** (either interval or ratio), the histogram is the natural generalization of a bar chart.  Histograms show frequency counts as well, but do so by lumping together ranges in the numerical data (the ranges are typically chosen automatically, i.e., to linearly span over the entire range of the data set).  We can construct a histogram using the `sns.distplot` command as in the following code.

### Important note: 
- Histograms are used to show distributions of variables while bar charts are used to compare variables
- Histograms plot quantitative data with ranges of the data grouped into bins or intervals while bar charts plot categorical data

```python
# Example 1 - Investigating the distribution of Revenue
x = raw_data['Revenue'].values # array with revenues only
print(x)
sns.distplot(x, color = 'blue'); # distplot stands for distribution plot

# As you can see, it's a bit imbalanced.
# Right skewed distribution as the mean is to the right
```

```python
# In Matplotlib
plt.hist(x, bins=15)
```

```python
# Example 2 - Investigating the distribution of Revenue, adding the mean

x = raw_data['Revenue'].values

sns.distplot(x, color = 'blue');

# Calculating the mean
mean = raw_data['Revenue'].mean()

#ploting the mean using matplotlib
plt.axvline(mean, 0,1, color = 'green') # axvline: Axis Vertical Line
```

```python
# Example 3 - Investigating the distribution of Visitors, adding the mean

x = raw_data['Visitors'].values

sns.distplot(x, color = 'red');

# Calculating the mean
mean = raw_data['Visitors'].mean()

#ploting the mean
plt.axvline(mean, 0,1, color = 'blue')
```

# 2D Data

## 1. Line plots â€” numeric x numeric (sequential)

For two dimensional data where one of the dimensions in naturally sequential (this comes up often, for instance, when monitoring a time series, so that one of the two dimensions of each data point is time).  

Because the sequential dimension is numeric (interval or ratio), in this case (unlike the bar charts), it _does_ make sense to explicitly draw lines between the data points.  Note that occasionally (if the data is sampled at regular intervals), this sometime considered to be one-dimensional data, but in this case the first dimension is an implicit sequential series, so we still consider the data to be two dimensional. 

The following examples illustrates how to use the `sns.lineplot` for a simple line plot to investigate the Revenue by Date.

```python
# ax = sns.lineplot(x='Date', y='Revenue', data = raw_data)
ax = sns.lineplot(x='Week_ID', y='Revenue', data = raw_data, ci=False, hue='Promo')
# Notes: error bands show the confidence interval
```

```python
grouped_df=raw_data.groupby(['Promo'])
# for key, item in grouped_df:
#     print(key, ":\n", grouped_df.get_group(key), "\n=======\n\n")
```

```python
# with matplotlib
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(1,1,1)
# or
# fig, ax = plt.subplots(figsize=(8,6))

for label, df in raw_data.groupby(['Promo']):
    df.plot(x='Week_ID', y='Revenue', ax=ax, label=label)
    
    
# raw_data.groupby(['Promo']).plot(x='Week_ID', y='Revenue')
# OR

# raw_data.groupby(['Promo']).plot(x='Week_ID', y='Revenue', ax=plt.gca())
# plt.legend()
# plt.xlabel('Week_ID')
# plt.ylabel('Revenue')
```

```python
# Example 2 sns - Adding Categories
# By Promo
ax = sns.lineplot(x='Week_ID', y='Revenue', data = raw_data, hue = 'Promo', ci=False)
```

```python
# Example 3 - By Promo with style
ax = sns.lineplot(x='Week_ID', y='Revenue', data = raw_data, hue = 'Promo',style='Promo', ci=False)
# In seaborn, the hue parameter determines which column in the dataframe should be used 
# for colour encoding
# Adding `hue="promo"` tells seaborn you want to colour 
# the data points for different promo categories differently
```

```python
# Example 4 - By Promo with style & Increase the size & Remove the confidence interval

# increase the size
sns.set(rc={'figure.figsize':(12,10)})
ax = sns.lineplot(x='Week_ID', y='Revenue', hue = 'Promo', style = 'Promo', data = raw_data, ci=None)

```

## 2. Scatter plots â€” numeric x numeric

If both dimensions of the data are numeric, the most natural first type of plot to consider is the scatter plot: plotting points that simply correspond to the different coordinates of the data.

Scatter plots are used to plot data points on a horizontal and a vertical axis in the attempt to show how much one variable is affected by another. 

The relationship between two variables is called their correlation.

We use scatter plots when we try to determine whether the two variables are related, such as: When trying to identify potential root causes of problems.

## 3. lmPlots â€” numeric x numeric

The lineplot `lmplot` is one of the most basic plots that shows a line on a 2 dimensional plan.
It is commonly used to show the results of a linear regression within a dataset.

An lmplots is a scatterplot with a regression line drawn on top of it.


- **What is Linear Regression**: 
It is a predictive statistical method for modelling the relationship between x (independent variable) & y (dependent V).
- [How it works?](https://towardsdatascience.com/machine-learning-fundamentals-via-linear-regression-41a5d11f5220)


```python
# Example 1 - Relationship between Marketing Spend and Revenue

ax = sns.lmplot(x="Marketing Spend", y="Revenue", data=raw_data, height=9, ci=False, line_kws={'color': 'red'})
```

**Interpretation**:
- This linear line across our plot is the best available fit for the trend of the revenue (y) with respect to the marketing spend made (x). 
- The data points that we see at extreme top right which are far away from this line are known as outliers in the dataset. We may think of outliers as exceptions.
- This line could be also used to predict the revenue based on the marketing spend

```python
# Example 2 - Relationship between Marketing Spend and Revenue - changing color, hue & Style

ax = sns.lmplot(x="Marketing Spend", y="Revenue", data=raw_data, height=9, ci=False, hue='Promo', markers=["o", "x", "+"])
```

**Interpretation**:

- Promotion Green has higher correlation between revenue and marketing spend. Promotion blue performs better in terms of revenue.
- For the other two categories (Promotion Orange and no promo) the correlation is there, but they do not perform as well as the other ones. The regression for no promo

```python
# breaking them down into different graphs

ax = sns.lmplot(x="Marketing Spend", y="Revenue", data=raw_data, col = 'Promo', ci= False, height=5, 
                line_kws={'color': 'red'}, 
                scatter_kws={'color':'#33FFDA'})
```

## 4. Heatmaps â€” categorical x categorical

When both dimensions of our 2D data are categorical, we have even less information to use.  Again, the goal will be to provide some kind of indication about the total counts of every possible combination between the two data sets. 

Heatmaps are commonly used to show relationships between two categorical variables, one plotted on each axis. By observing how cell colors change across each axis, you can observe if there are any patterns in value for one or both variables.

More details [here](https://chartio.com/learn/charts/heatmap-complete-guide/#:~:text=Heatmaps%20are%20used%20to%20show,for%20one%20or%20both%20variables.)

- First we need to create a "Dataset" to display on a Heatmap, this dataset could be a confusion matrix, correlation matrix, etc.
- In this example, we will use a correlation dataset
- .corr() is used to find the pairwise correlation of all columns in the dataframe. Any null values are automatically excluded
- The closer to 1 or -1 the better. As one variable increases, the other variable tends to also increase / decrease
- More Info [here](https://statisticsbyjim.com/basics/correlations/)

#### Example 1 - Heatmap for Pearson Correlation


```python
# First Calculate the pairwise correlation for the columns containing numeric values and save it in a dataframe
pc_df = raw_data[['Revenue','Visitors','Marketing Spend']].corr(method ='pearson')
pc_df
```

```python
ax = sns.heatmap(pc_df)
```

**Interpretation**
- The correlation scale to the right does not start from -1, but rather from 0.57, this is because the lowest correlation value we got is 0.573842.
- The darker the color is the lower the correlation
- Lighter colors correspond to higher correlation
- the number of `visitors` does not affect much the revenue, the `revenue` is rather affected by the `marketing spend`

```python
# rename the columns
cols = ['Revenue PC','Visitors PC','Marketing Spend PC']

ax = sns.heatmap(pc_df, annot=True,
                 yticklabels=cols,
                 xticklabels=cols,
                 annot_kws={'size': 10})
```

```python

# Example 2 - Heatmap for PC - changing cmap

ax = sns.heatmap(pc_df, annot=True,
                 yticklabels=cols,
                 xticklabels=cols,
                 annot_kws={'size': 30},
                 cmap="Greens")

# Examples:
# cmap="YlGnBu"
# cmap="Blues"
# cmap="BuPu"
# cmap="Greens"
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(raw_data.corr(),annot=True,fmt='.1f')
plt.show()
```

# Holistic views / Combo:

## 1. SubPlots

```python
# Set up the matplotlib figure
fig, axes = plt.subplots(2, 2, figsize=(12, 7))

a = raw_data['Revenue'].values
b = raw_data['Visitors'].values
c = raw_data['Marketing Spend'].values

# plot 1
sns.distplot(a, color = 'blue', ax=axes[0,0])

# plot 2
sns.distplot(b, color = 'blue', ax=axes[0,1])

# plot 3
sns.distplot(c, color = 'blue', ax=axes[1,0])

# plot 4
sns.boxplot(x="Revenue", y="Day_Name", data=raw_data, color = '#52F954', ax=axes[1,1])
sns.swarmplot(x="Revenue", y="Day_Name", data=raw_data, color="red", ax=axes[1,1])
#A swarm plot can be drawn on its own, but it is also a good complement to a box or violin plot in cases where you want to show all observations along with some representation of the underlying distribution.
```
